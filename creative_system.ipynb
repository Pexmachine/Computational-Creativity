```python
def lookup_ingredient(ingredient):
    for i in data:
        if i["entity_alias_readable"] == ingredient:
            return i
        

def ingredient_category(ingredient):
    ingr = lookup_ingredient(ingredient)
    
    return ingr["category"]
    


def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3

def count_shared_flavor_profiles(ingr1,ingr2):
    ingredient1 = lookup_ingredient(ingr1)
    ingredient2 = lookup_ingredient(ingr2)
    
    return len(intersection(ingredient1["molecules"],ingredient2["molecules"]))

def Sort_Tuple(tup): 
  
    # reverse = None (Sorts in Ascending order) 
    # key is set to sort using second element of 
    # sublist lambda has been used 
    return(sorted(tup, key = lambda x: x[1], reverse = True))


def remove_same_combinations(list):
    newList = list
    
    for combination in list:
        if (combination[1],combination[0],combination[2]) in newList:
            newList.remove(combination)
    return newList

def count_shared_profiles(list):
    sum = 0
    
    for combination in list:
        sum += combination[2]
    
    return sum

def count_overall_shared_profiles(recipe):
    shared_flavor_profiles = []
    
    for ingredient1 in recipe:
        for ingredient2 in recipe:
            if ingredient1 != ingredient2:
                shared_flavor_profiles.append((ingredient1,ingredient2,count_shared_flavor_profiles(ingredient1,ingredient2)))
                
    remove_duplicates = list(dict.fromkeys(shared_flavor_profiles))
    
    remove_same_combis = remove_same_combinations(remove_duplicates)
    
    return count_shared_profiles(remove_same_combis)

def shared_with_recipe(recipe,ingredient):
    sum = 0
    
    for recipeingredient in recipe:
        sum += count_shared_flavor_profiles(recipeingredient,ingredient)
    
    return sum

 

def ranked_list_of_shared_flavor_profiles(ingredient):
    shared_flavor_profiles = map(lambda x: [x,count_shared_flavor_profiles(ingredient,x)], available_ingredients)
    
    rankedl = Sort_Tuple(shared_flavor_profiles)
    
    list_without_ingredient = filter (lambda c : not c[0] == ingredient, rankedl)
    
    return list(list_without_ingredient)

def rl_around_recipe(recipe):
    rl = map(lambda x: [x,shared_with_recipe(recipe,x)],available_ingredients)
    
    ranked1 = Sort_Tuple(rl)
    
    list_without_recipe_ingredients = filter (lambda c : c[0] not in recipe, ranked1)
    
    
    return list(list_without_recipe_ingredients)


def is_complete_meal(recipe):
    contains_meat = False
    contains_vegetable = False
    
    for ingredient in recipe:
        if ingredient_category(ingredient) == 'meat':
            contains_meat = True
        elif ingredient_category(ingredient) == 'vegetable':
            contains_vegetable = True
            
    return contains_meat and contains_vegetable

def has_meat(recipe):
    contains_meat = False
    
    for ingredient in recipe:
        if ingredient_category(ingredient) == 'meat':
            contains_meat = True
    
    return contains_meat


def has_vegetable(recipe):
    contains_vegetable = False
    
    for ingredient in recipe:
        if ingredient_category(ingredient) == 'vegetable':
            contains_vegetable = True
    
    return contains_vegetable

"""
print(count_overall_shared_profiles(new))
print(shared_with_recipe(['Honey','Beef','Mushroom','Basmati Rice'],'Apple'))

print(ingredient_category('Apple'))

"""
print(ranked_list_of_shared_flavor_profiles('Apple'))
```
